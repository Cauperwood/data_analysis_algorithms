```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

1. Подберите скорость обучения (alpha) и количество итераций:

```python
def plot_gradient_descent(err_hist: np.array, W_history: np.array):
    
    
    plt.figure(figsize=(13, 6))
    plt.plot(err_hist, linewidth=3)
    plt.title('Gradient descent')
    plt.xlabel('err')
    plt.ylabel('iterations')
    
    
    plt.figure(figsize=(13, 6))
    plt.plot(W_history, linewidth=3)
    plt.title('WEITHS')
    plt.xlabel('WEITHS')
    plt.ylabel('iterations')
    
    
    plt.show()
```
